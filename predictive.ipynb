```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import datetime

plt.style.use('cbi')
```

```python
df = pd.read_csv('games/regseason05-14 final.csv')
df.Home = df.Home != False
```

```python
y = df.Year.unique()[0]

for t in sorted(df.Tm.unique()):
    minyr = df[df.Tm == t].Year.min()
    maxyr = df[df.Tm == t].Year.max()
    if not ((minyr == 2005) & (maxyr == 2014)):
        pass
        # print "%s: %d-%d" % (t,minyr,maxyr)

# Update TBD --> TBR
df.loc[df.Tm == 'TBD', 'Tm'] = 'TBR'
df.loc[df.Opp == 'TBD', 'Opp'] = 'TBR'
# Update FLA --> MIA
df.loc[df.Tm == 'FLA', 'Tm'] = 'MIA'
df.loc[df.Opp == 'FLA', 'Opp'] = 'MIA'

df.insert(7, 'WinId', df['W/L'].apply(lambda w: "W" in w))
df['Wcount'] = 0
df['Rsum'] = 0
df['RAsum'] = 0
df.insert(9, 'RD', df.R - df.RA)

for y in df.Year.unique():
    tms = df[df.Year == y].Tm.unique()
    for tm in tms:
        df.loc[(df.Year == y) & (df.Tm == tm) ,'Wcount'] = df.loc[(df.Year == y) & (df.Tm == tm) ,'WinId'].cumsum()
        df.loc[(df.Year == y) & (df.Tm == tm) ,'Rsum'] = df.loc[(df.Year == y) & (df.Tm == tm) ,'R'].cumsum()
        df.loc[(df.Year == y) & (df.Tm == tm) ,'RAsum'] = df.loc[(df.Year == y) & (df.Tm == tm) ,'RA'].cumsum()
        
# df.drop(['box', 'Rank', 'Time', 'Save', 'Attendance', 'W/L', 'W-L', 'Inn', 'Streak', 'Win', 'Loss'], axis = 1, inplace = True)
df.insert(14, 'Lcount', df['Gm#'] - df.Wcount)        
df['RDsum'] = df.Rsum - df.RAsum
df['PythWL'] = df.apply(lambda r: pow(r.Rsum, 1.81)/(pow(r.Rsum, 1.81) + pow(r.RAsum, 1.81)), axis = 1)
```

```python
df.drop(['Rank'], axis = 1, inplace = True)
```

```python
df2 = df.merge(df[['Tm', 'Date', 'Year', 'RDsum', 'PythWL']], 
               how = 'left',
               left_on = ['Opp', 'Date', 'Year'],
               right_on = ['Tm', 'Date', 'Year'],
               suffixes = ['', '_def']
               )

df2.drop('Tm_def', axis = 1, inplace = True)
df2.Home = df2.Home.map(int)
df2.WinId = df2.WinId.map(int)
```

```python
df2['RDAvg'] = df2.RDsum/df2['Gm#']
df2['RDAvg_def'] = df2.RDsum_def/df2['Gm#']
```
