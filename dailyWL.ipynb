```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
df = pd.read_csv('games/regseason05-14 final.csv')
```

```python
y = df.Year.unique()[0]

for t in sorted(df.Tm.unique()):
    minyr = df[df.Tm == t].Year.min()
    maxyr = df[df.Tm == t].Year.max()
    if not ((minyr == 2005) & (maxyr == 2014)):
        print "%s: %d-%d" % (t,minyr,maxyr)
        
# Update FLA --> MIA
# Update TBD --> TBR
```

```python
df2 = df.copy()

df2.loc[df2.Tm == 'TBD', 'Tm'] = 'TBR'
df2.loc[df2.Opp == 'TBD', 'Opp'] = 'TBR'

df2.loc[df2.Tm == 'FLA', 'Tm'] = 'MIA'
df2.loc[df2.Opp == 'FLA', 'Opp'] = 'MIA'

df2.insert(8, 'WinId', df2['W/L'].apply(lambda w: "W" in w))
df2['Wcount'] = 0
df2['Rsum'] = 0
df2['RAsum'] = 0
```

```python
# for y in df.Year.unique():df
y = 2005
```

```python
for y in df2.Year.unique():
    tms = df2[df2.Year == y].Tm.unique()
    for tm in tms:
        df2.loc[(df2.Year == y) & (df2.Tm == tm) ,'Wcount'] = df2.loc[(df2.Year == y) & (df2.Tm == tm) ,'WinId'].cumsum()
        df2.loc[(df2.Year == y) & (df2.Tm == tm) ,'Rsum'] = df2.loc[(df2.Year == y) & (df2.Tm == tm) ,'R'].cumsum()
        df2.loc[(df2.Year == y) & (df2.Tm == tm) ,'RAsum'] = df2.loc[(df2.Year == y) & (df2.Tm == tm) ,'RA'].cumsum()
```

```python
df_gmerr = df2.groupby('Gm#').agg({'PythErrAbs' : ['mean', 'median', 'std']})
df_gmerr['min'] = df_gmerr['PythErrAbs']['mean'] - df_gmerr['PythErrAbs']['std']
df_gmerr['max'] = df_gmerr['PythErrAbs']['mean'] + df_gmerr['PythErrAbs']['std']


fig, ax = plt.subplots(1,1, figsize = (10,6))
# ax.plot(df2.groupby('Gm#').agg({'PythErrAbs' : ['mean', 'median']})['PythErrAbs']['median'].values);
# ax.plot(pd.rolling_mean(df2.groupby('Gm#')['PythErrAbs'].mean().values[:162], 3))
ax.plot(df2.groupby('Gm#')['PythErrAbs'].mean().values[:163])
ax.scatter(162, df2.groupby('Gm#')['PythErrAbs'].mean().values[162])

# ax.plot(pd.rolling_mean(df_gmerr['min'], 3), linestyle='--', color='r')
# ax.plot(pd.rolling_mean(df_gmerr['max'], 3), linestyle='--', color='r')
ax.fill_between(range(163), 
                 df_gmerr['min'], 
                 df_gmerr['max'], 
                 facecolor='red', 
                 alpha=0.25)


ax.set_xlim([0,170])
ax.set_ylim([0,0.25])

ax.set_xticks(range(0,161,40))
ax.set_yticklabels(ax.get_yticks()*162)

ax.set_xlabel('Game #')
ax.set_ylabel('Pyth Error (Games)')

plt.annotate(
        round(df2.groupby('Gm#').agg({'PythErrAbs' : ['mean', 'median']})['PythErrAbs']['median'][162]*162,1),
        xy = (162, df2.groupby('Gm#').agg({'PythErrAbs' : ['mean', 'median']})['PythErrAbs']['median'][162]), xytext = (20,15),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(fc = 'blue', alpha = 0.25),
        arrowprops = dict(arrowstyle = '-',))

plt.annotate(
        round(df2.groupby('Gm#').agg({'PythErrAbs' : ['mean', 'median']})['PythErrAbs']['median'][162]*162,1),
        xy = (162, df2.groupby('Gm#').agg({'PythErrAbs' : ['mean', 'median']})['PythErrAbs']['median'][162]), xytext = (20,15),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(fc = 'blue', alpha = 0.25),
        arrowprops = dict(arrowstyle = '-',))

plt.suptitle('Mean Pythagorean Error by Game, 2005-2014', fontsize = 18);
```

```python
fig, axes = plt.subplots(2, 5, figsize = (20,6))
for ax, yr in zip(axes.flat, df2.Year.unique()[-10:]):
    df_gmerr_sea = df2[df2.Year == yr].groupby('Gm#').agg({'PythErrAbs' : ['mean', 'median', 'std']})
    ax.plot(df_gmerr_sea['PythErrAbs']['mean'].values);
    
    df_gmerr_sea['min'] = df_gmerr_sea['PythErrAbs']['mean'] - df_gmerr_sea['PythErrAbs']['std']
    df_gmerr_sea['max'] = df_gmerr_sea['PythErrAbs']['mean'] + df_gmerr_sea['PythErrAbs']['std']
    
    ax.fill_between(range(len(df_gmerr_sea['PythErrAbs']['mean'])), 
                 df_gmerr_sea['min'], 
                 df_gmerr_sea['max'], 
                 facecolor='red', 
                 alpha=0.25)
    
    ax.set_ylim([0,0.25])
    ax.set_xlim([0,170])
    
    ax.set_xticks(range(0,161,40))
    ax.set_yticklabels(ax.get_yticks()*162)
    
    ax.set_xlabel('Game #') 
    ax.set_title(yr)

plt.suptitle('Pythagorean Error by Game, by Season', fontsize = 18, y = 1.05)
plt.tight_layout();
```

```python
Correlation between Game n Pyth and end of season winP
pyth_corrs= []
for tm in tms[:5]:
    for yr in df2.Year.unique()[-10:]:
        # yr = 2005
        dfty = df2[(df2.Tm == tm) & (df2.Year == yr)].sort('Gm#')
        max_games = dfty['Gm#'].max()
        final_wp = dfty.loc[dfty['Gm#'] == max_games, 'WinP'].values[0]
        for ii in range(1,max_games+1):
            pyth_corrs.append([yr, ii, dfty.loc[dfty['Gm#'] == ii, 'Pyth'].values[0], final_wp])
```

```python
dfpyth = pd.DataFrame(pyth_corrs, columns=['Year', 'Gm', 'WinP', 'Pyth'])
dfpyth.head()
```
